# Crelu Activation Function

## In Tensorflow :D

### With visualization ;)

Computes Concatenated ReLU.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

function_name ='Crelu'
```

### Data creation

Three toy examples (instances, observations) with 5 features created from random normal dist.

```python
numInstances = 3

data = []

for _ in range(numInstances):
    instance = np.random.normal(size=5)
    data.append(instance)
    
data
```

### Placeholder & The Op

```python
data_ph = tf.placeholder(dtype=tf.float32,shape=[3,5])
```

```python
relu = tf.nn.crelu(data_ph)
```

### Execute graph

```python
with tf.Session() as sess:
    
    _relu = sess.run(relu,feed_dict={data_ph:data})

print('Relu output: (%d,%d) <=> (instances,features)\n%s' % (_relu.shape[0],_relu.shape[1],_relu))
```

### Visualize

```python
x_values = np.arange(-10,10,0.1)
dynSess = tf.Session()
y_values = dynSess.run(tf.nn.crelu(x_values))
```

```python
plt.figure(figsize=(10,6))
plt.title(function_name)
plt.xlabel('input value')
plt.ylabel('output value')
plt.grid()
plt.plot(x_values,y_values[0:200])
plt.show();
```

```python
plt.figure(figsize=(10,6))
plt.title(function_name)
plt.xlabel('input value')
plt.ylabel('output value')
plt.grid()
plt.plot(x_values,y_values[200:400])
plt.show();
```
