# Dropout Activation Function

## In Tensorflow :D

### With visualization ;)

Computes dropout.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

function_name ='Dropout'
```

### Data creation

Three toy examples (instances, observations) with 5 features created from random normal dist.

```python
numInstances = 3

data = []

for _ in range(numInstances):
    instance = np.random.normal(size=5)
    data.append(instance)
    
data
```

### Placeholder & The Op

```python
data_ph = tf.placeholder(dtype=tf.float32,shape=[3,5])
```

```python
dropout = tf.nn.dropout(data_ph,keep_prob=0.5)
```

### Execute graph

```python
with tf.Session() as sess:
    
    _dropout = sess.run(dropout,feed_dict={data_ph:data})

print('Relu output: (%d,%d) <=> (instances,features)\n%s' % (_dropout.shape[0],_dropout.shape[1],_dropout))
```

### Visualize

```python
x_values = np.arange(-10,10,0.1)
dynSess = tf.Session()
y_values = dynSess.run(tf.nn.dropout(x_values,keep_prob=0.5))
```

```python
plt.figure(figsize=(10,6))
plt.title(function_name)
plt.xlabel('input value')
plt.ylabel('output value')
plt.grid()
plt.plot(x_values,y_values)
plt.show();
```
