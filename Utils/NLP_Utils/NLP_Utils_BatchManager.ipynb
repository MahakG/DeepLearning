# Batch Manager demonstration

BatchManager is a class inside the NLP_Utils package

```python
import NLP_Utils as nlp
```

## Dense

```python
mode = 'dense'
```

```python
filename = 'dummy.txt'
batch_size = 2
seq_length = 20
overlap = 18
vocab_strategy='all'
```

```python
batchManager = nlp.BatchManager()
```

```python
batchManager.get_params()
```

```python
batchManager.set_params(filename=filename,batch_size=batch_size,seq_length=seq_length,overlap=overlap,
                        vocab_strategy=vocab_strategy,mode=mode)
```

```python
batchManager.get_params()
```

```python
batchManager.stats()
```

```python
index = 1

for batchX,batchY,epoch in batchManager.generate_batches():
    
    batchX = np.array(batchX)
    batchY = np.array(batchY)
    
    print('New batch (%d):' % index)
    print('Batch X shape:',batchX.shape,'||','Batch Y shape:',batchY.shape)
    print('(X:1): ',batchX[0],'=>',batchManager.vocab_decode(batchX[0]))
    print('(Y:1): ',batchY[0],'=>',batchManager.vocab_decode(batchY[0]))
    print('(X:2): ',batchX[1],'=>',batchManager.vocab_decode(batchX[1]))
    print('(Y:2): ',batchY[1],'=>',batchManager.vocab_decode(batchY[1]))

    
    
    print()
    
    index += 1
    
```
